## Merge clinical and imaging features

Merging datasets before analysis with the agents. 

```python
import pandas as pd
```

```python
# import data - separate cells because imaging_features are large
imaging_feats = pd.read_excel('data/Imaging_Features.xlsx')
```

```python
clinical_feats = pd.read_excel('data/Clinical_and_Other_Features.xlsx', skiprows=1)
```

```python
# binarize dataset - clinical feats contains the molecular subtype. 
# "{0 = luminal-like, 1 = ER/PR pos, HER2 pos, 2 = her2, 3 = trip neg}. label 3 becomes our target variable (1). other labels are negative
# from gemini

clinical_feats['HRD_Proxy'] = (clinical_feats['Mol Subtype'] == 3).astype(int)

# Keep only the ID and your new target, then merge with the imaging features
# (Assuming the shared ID column is 'Patient ID')
merged_df = pd.merge(
    imaging_feats, 
    clinical_feats[['Patient ID', 'HRD_Proxy']], 
    on='Patient ID', 
    how='inner'
)

# Drop any rows with missing values to keep it clean and fast
merged_df = merged_df.dropna()

```

```python
# save dataset
merged_df.to_csv('data/merged_image_clinical_features.csv', index=False)
```
